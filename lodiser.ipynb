```python
# 1. Cargar los datos
detalle_ventas = pd.read_excel(r"C:\Users\primo\Desktop\Ciencia de Datos\Proyectos\Lodiser\detalle_ventas.xlsx")
productos = pd.read_excel(r"C:\Users\primo\Desktop\Ciencia de Datos\Proyectos\Lodiser\productos.xlsx")

# 2. Unir los datos
ventas_completo = detalle_ventas.merge(productos[['id_producto', 'categoría']], on='id_producto', how='left')

# 3. Calcular el total vendido por producto
ventas_por_producto = (
    ventas_completo
    .groupby(['id_producto', 'categoría'], as_index=False)
    .agg(total_venta=('subtotal', 'sum'))
)

# 4. Rankear de mayor a menor por categoría
ventas_por_producto['rank'] = (
    ventas_por_producto
    .groupby('categoría')['total_venta']
    .rank(method='dense', ascending=False)
)

# 5. También creamos el rank inverso (últimos puestos)
ventas_por_producto['rank_inverso'] = (
    ventas_por_producto
    .groupby('categoría')['total_venta']
    .rank(method='dense', ascending=True)
)

# 6. Unimos los rankings al DataFrame original
ventas_modificadas = ventas_completo.merge(
    ventas_por_producto[['id_producto', 'rank', 'rank_inverso']],
    on='id_producto',
    how='left'
)

# 7. Función para ajustar la cantidad
def ajustar_cantidad(row):
    if row['rank'] == 1:
        return row['cantidad'] + 3
    elif row['rank'] == 2:
        return row['cantidad'] + 2
    elif row['rank'] == 3:
        return row['cantidad'] + 1
    elif row['rank_inverso'] <= 2 and row['cantidad'] > 1:
        return row['cantidad'] - 1
    else:
        return row['cantidad']

# 8. Aplicar el ajuste
ventas_modificadas['cantidad_ajustada'] = ventas_modificadas.apply(ajustar_cantidad, axis=1)
```

```python
# 9. Recalcular subtotal en base a la cantidad ajustada
ventas_modificadas['subtotal_ajustado'] = ventas_modificadas['cantidad_ajustada'] * ventas_modificadas['precio_unitario']
ventas_modificadas.head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Ruta del archivo de ventas
ruta_ventas = r"C:\Users\primo\Desktop\Ciencia de Datos\Proyectos\Lodiser\ventas.xlsx"

# Leer archivo
ventas = pd.read_excel(ruta_ventas)

# Asegurarse que la columna de fecha esté en formato datetime
ventas['fecha'] = pd.to_datetime(ventas['fecha'])

# Filtrar solo entre enero y abril 2025
ventas = ventas[(ventas['fecha'] >= '2025-01-01') & (ventas['fecha'] <= '2025-04-30')]

# Agregar columnas auxiliares
ventas['mes'] = ventas['fecha'].dt.month
ventas['año'] = ventas['fecha'].dt.year
ventas['semana'] = ventas['fecha'].dt.to_period('W').apply(lambda r: r.start_time)

# Agrupar por semana original
ventas_semanales = ventas.groupby('semana', as_index=False)['total_venta'].sum()

# Agregar columna de mes para poder redistribuir
ventas_semanales['mes'] = ventas_semanales['semana'].dt.month

# Función para redistribuir ventas manteniendo total mensual
def redistribuir_mes(df_mes, dispersion=0.08):
    semanas = len(df_mes)
    factores = np.random.normal(loc=1.0, scale=dispersion, size=semanas)
    factores = np.maximum(factores, 0.1)  # evitar negativos o muy bajos
    factores /= factores.sum()  # normalizar
    total_mes = df_mes['total_venta'].sum()
    nuevos_totales = factores * total_mes
    df_mes['total_venta'] = nuevos_totales
    return df_mes

# Aplicar redistribución por mes
ventas_suavizadas = pd.concat([
    redistribuir_mes(grupo) for _, grupo in ventas_semanales.groupby('mes')
])

# Graficar para verificar
plt.figure(figsize=(12, 5))
sns.lineplot(data=ventas_suavizadas, x='semana', y='total_venta', marker='o')
plt.title("Ventas semanales suavizadas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
